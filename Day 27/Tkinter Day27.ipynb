```python
##Using pack() to display items on the canvas



import tkinter
window=tkinter.Tk()
window.minsize(width=500,height=300)
window.title("Button click")


#button click event listner
def fun_click():
    my_label["text"]="button clicked"
    my_label2["text"]=inp.get()

#label
my_label=tkinter.Label(text="label")
my_label.pack()

#label2
my_label2=tkinter.Label(text="label2")
my_label2.pack()


#button
my_button=tkinter.Button(text="button",command=fun_click)
my_button.pack()


#input
inp=tkinter.Entry(width=10)
inp.pack()


window.mainloop()
```

```python
##Using grid() to display items on the canvas


import tkinter
window=tkinter.Tk()
window.minsize(width=500,height=300)
window.title("Button click")


#button click event listner
def fun_click():
    my_label["text"]="button clicked"
    my_label2["text"]=inp.get()

#label
my_label=tkinter.Label(text="label")
my_label.grid(column=1,row=1)

#label2
my_label2=tkinter.Label(text="label2")
my_label2.grid(column=3,row=1)


#button
my_button=tkinter.Button(text="button",command=fun_click)
my_button.grid(column=2,row=2)


#input
inp=tkinter.Entry(width=10)
inp.grid(column=4,row=4)


window.mainloop()
```

```python
## Miles to KM project using tkinter
import tkinter
window=tkinter.Tk()
window.title("Mile to Km Converter")



#calculate function (eventlistner)
def calFunction():
    miles=miles_input.get()
    km_number["text"]=((int)(miles))*1.609

#adding the first entry (to inter the miles) from the user
miles_input=tkinter.Entry(width=8)

#adding Miles label next to miles input
miles_label=tkinter.Label(text="Miles")

#Km label which will be empty at start but will be populated with km value when calculate is clicked
km_number=tkinter.Label(text="")

#Km label with Km written as the unit
km_label=tkinter.Label(text="Km")

#is equal to label
is_equal_to=tkinter.Label(text="is equal to ")

#calculate button
cal=tkinter.Button(text="Calculate",command=calFunction)


#placing everything on the canvas

is_equal_to.grid(row=2,column=1)
miles_input.grid(row=1,column=2)
miles_label.grid(row=1,column=3)
km_number.grid(row=2,column=2)
km_label.grid(row=2,column=3)
cal.grid(row=3,column=2)



window.mainloop()
```
