```python
#Day 28 Making a pomodoro application using tkinter gui

import tkinter
import math
# ---------------------------- CONSTANTS ------------------------------- #
PINK = "#e2979c"
RED = "#e7305b"
GREEN = "#9bdeac"
YELLOW = "#f7f5dd"
FONT_NAME = "Courier"
WORK_MIN = 25
SHORT_BREAK_MIN = 5
LONG_BREAK_MIN = 20
rep=0
timer=None

# ---------------------------- TIMER RESET ------------------------------- # 

def reset():
    global rep
    rep=0
    window.after_cancel(timer)
    timer_label["text"]="Timer"
    canvas.itemconfig(timer_text,text=('00:00'))
    tick.config(text="")
# ---------------------------- TIMER MECHANISM ------------------------------- # 

def start_timer():
    global rep
    timer_time=1
    rep+=1
    work_rep=[1,3,5,7]
    rest_rep_small=[2,4,6]
    if rep in work_rep:
        timer_time=60*WORK_MIN
        timer_label.config(text="Work",fg=GREEN)
    elif rep in rest_rep_small:
        timer_time=60*SHORT_BREAK_MIN
        timer_label.config(text="Break",fg=PINK)
    elif rep==8:
        timer_time=60*LONG_BREAK_MIN
        timer_label.config(text="Break",fg=RED)
        
    
    count_down(timer_time)

# ---------------------------- COUNTDOWN MECHANISM ------------------------------- # 

def count_down(count):
    count_min=math.floor(count/60)
    count_sec=count%60
    
    if count_sec==0:
        count_sec="00"
    elif count_sec<10:
        count_sec="0"+str(count_sec)
    
    canvas.itemconfig(timer_text,text=f'{count_min}:{count_sec}')
    if count>0:
        global timer
        timer=window.after(1000,count_down,count-1)
    else:
        
        text="✔ "
        if rep in [2,4,6,8]:
            tick.config(text=text*int(rep/2))
        start_timer()

# ---------------------------- UI SETUP ------------------------------- #
window=tkinter.Tk()
window.title("Pomodoro")
window.config(padx=50,pady=50,bg=YELLOW)


#adding the image to the canvas using tkinter canvas widget
canvas=tkinter.Canvas(width=200,height=225,bg=YELLOW,highlightthickness=0) 
# this is based on the size of image we have

tomato_img=tkinter.PhotoImage(file="tomato.png")
canvas.create_image(100,112,image=tomato_img)
timer_text=canvas.create_text(103,130,text="00:00",fill="white",font=(FONT_NAME,35,"bold"))
canvas.grid(row=2,column=2)


#adding timer title on the top
timer_label=tkinter.Label(text="Timer")
timer_label.config(fg=GREEN,bg=YELLOW,font=(FONT_NAME,45))
timer_label.grid(row=1,column=2)

#adding start button
start_button=tkinter.Button(text="Start",highlightthickness=0,command=start_timer)
start_button.config(bg=YELLOW)
start_button.grid(row=3,column=1)



#adding reset button

reset_button=tkinter.Button(text="Reset",highlightthickness=0,command=reset)
reset_button.config(bg=YELLOW)
reset_button.grid(row=3,column=3)


#adding a tick 
tick=tkinter.Label(text="")
tick.config(bg=YELLOW,fg=GREEN)
tick.grid(row=4,column=2)

window.mainloop()
```
