```python
import networkx as nx
import matplotlib.pyplot as plt
import random

plt.figure(figsize=(15,15),dpi=200)
N=50 #number of nodes in the graph

#creating a graph with n nodes 1 to N
G=nx.Graph()
G.add_nodes_from([i for i in range(1,N+1)])

#creating weighted edges to make the graph complete
edge_weight=["+","-"] #here + represent that node 1 and node 2 is friend and - represent enimies

for i in G.nodes():
    for j in G.nodes():
        if i!=j:
            G.add_edge(i,j,sign=random.choice(edge_weight))
            


#display the graph initially

edge_labels=nx.get_edge_attributes(G,'sign')
pos=nx.circular_layout(G)
nx.draw(G,pos=pos,node_size=2000,with_labels=1,font_size=30)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels,font_size=30,font_color="red")
plt.show()


#getting the list of all nodes
nodes=G.nodes()

#all possible trianges
ti
```
